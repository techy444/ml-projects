```python
import pandas as pd
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
digits
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.gray() #This sets the colormap to grayscale.
for i in range(4):
    plt.matshow(digits.images[i])# used to display a 2D matrix as an image
```

```python
df=pd.DataFrame(digits.data)
```

```python
df.head()
```

```python
df['target']=digits.target
```

```python
df.head()
```

```python
#split dataset into independent and dependent features
X=df.iloc[:,:-1]##this will take all the columns except last once and these are independent features
y=df.iloc[:,-1]#dependent features
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
len(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
ypred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_test,ypred)
```

```python
print(classification_report(y_test,ypred))
```

```python
from xgboost import XGBClassifier

xgmodel=XGBClassifier()

xgmodel.fit(X_train,y_train)
y_pred = xgmodel.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
