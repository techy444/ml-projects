```python
import seaborn as sns 
import pandas as pd
import numpy as np
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
df['species']=df['species'].map({'versicolor':0,'virginica':1,'setosa':2})
```

```python
df.head()
```

```python
#split dataset into independent and dependent features
X=df.iloc[:,:-1]##this will take all the columns except last once and these are independent features
y=df.iloc[:,-1]#dependent features
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
#preprunning
parameters={
    'criterion':['gini', 'entropy', 'log_loss'], 
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt', 'log2']
}
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
model=DecisionTreeClassifier()
cv=GridSearchCV(model,param_grid=parameters,cv=5,scoring='accuracy')
```

```python
cv.fit(X_train,y_train)
```

```python
cv.best_params_
```

```python
ypred=cv.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(ypred,y_test)
```

```python
print(score)
```

```python
print(classification_report(y_test,ypred))
```
